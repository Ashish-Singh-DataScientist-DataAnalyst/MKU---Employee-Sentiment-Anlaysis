```python
def rename_columns(df, new_names_list):
    
    """Takes a DataFrame that needs to be renamed and a list of the new
        column names, and returns the renamed DataFrame. Make sure the 
        number of columns in the df matches the list length exactly,
        or function will not work as intended."""
    
    rename_dict = dict(zip(df.columns, new_names_list))
    df = df.rename(mapper=rename_dict, axis=1)
    
    return df
```

```python
headers[61:65].values
```

```python
ic_col_names = headers[61:65].values

important_consideration6 = rename_columns(important_consideration6, ic_col_names)

# Now tack on age_group from the original DataFrame so we can use .groupby
# (You could also use pd.concat, but I find this easier)
important_consideration6['What is your age?'] = df['What is your age?']
important_consideration6['How long have you worked for this organization?'] = df['How long have you worked for this organization?']
important_consideration6['Department Coding'] = df['Department Coding']
important_consideration6['Designation Coding'] = df['Designation Coding']
important_consideration6['Pay & Benefits'] = df['Pay & Benefits']
important_consideration6.head(20)
```

```python
important_consideration6.columns
```

```python
mycode = {'Neutral' : 0, 'Disagree Strongly': 1, 'Somewhat Disagree' : 2, 'Agree' : 3, 'Agree Somewhat' : 4, 'Agree Strongly':5 }
```

```python
for i in important_consideration6.columns:
    important_consideration6[i] = important_consideration6[i].replace(mycode)
      
```
